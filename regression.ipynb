```python
# Bring our packages in 
import pandas as pd
from scipy import stats
import statsmodels.formula.api as smf

%matplotlib inline 
```

```python
# Connect to our data

myRegressionData = pd.read_csv("C:\\Users\\Malee\\OneDrive\\Desktop\\Self Learning\\The Data Science of Marketing (LinkedIn)\\Exercise Files\\03_03\\regression-py.csv")
```

```python
# View a snapshot of our data 
myRegressionData
```

```python
# Plot the data
myRegressionData.plot(kind='scatter', x='broadcast', y='sales')
```

```python
# Calculate r-squared
slope, intercept, r_value, p_value, std_err = stats.linregress(myRegressionData.broadcast,myRegressionData.sales)
```

```python
# Print the r-squared value
print("R-Squared: {}".format(r_value**2))
#The result shows a very strong correlation between broadcast and sales
```

```python
# Model OLS to generate coefficients
myLinearModel = smf.ols(formula = 'sales ~ broadcast', data=myRegressionData).fit()
```
